```python
import random
import numpy as np
import matplotlib.pyplot as plt
lista=[]
```

```python
def expo(lamda,n):
  L=-1/lamda
  for i in range(n):
    u=random.random()
    x=L*np.log(u)
    lista.append(x)
  return lista
```

```python
G=expo(0.5,10000)
mean=np.mean(G)
variance=np.var(G)
```

```python
plt.hist(G)
plt.show()
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(G)
```

```python
plt.step(X,Y)
```

```python
plt.plot(X,Y)
plt.show()
```
