```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Initialize variables
lista = []
cont = 0
N = 16

# Iterate N times
for i in range(N):
    u1 = random.random()
    u2 = random.random()

    # Check if (u1, u2) is under the quarter-circle curve
    if u2 <= np.sqrt(1 - u1**2):
        cont += 1
        plt.scatter(u1, u2, color="red")
    else:
        plt.scatter(u1, u2, color="blue")

    # Plot the quarter-circle curve
    x = np.linspace(0, 1, 100)
    plt.plot(x, np.sqrt(1 - x**2), color="green")  # Fixed the color and indentation

    # Calculate and append the probability
    p = cont / (i + 1)  # Probability after i+1 iterations
    lista.append(p)

# Show the plot
plt.axis('equal')
plt.show()

# Print the final list of probabilities
print("List of probabilities:", lista)
```
