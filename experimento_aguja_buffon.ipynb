```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del experimento
L = 1.0  # Longitud de la aguja
D = 1.0  # Distancia entre las lneas
num_trials = 10000  # Nmero de lanzamientos

def throw_needle(L, D):
    # Generar un ngulo aleatorio entre 0 y /2
    theta = np.random.uniform(0, np.pi/2)
    # Generar una distancia aleatoria del centro de la aguja a la lnea ms cercana
    y = np.random.uniform(0, D/2)
    # Determinar si la aguja cruza una lnea
    if y <= (L/2) * np.sin(theta):
        return 1
    else:
        return 0

# Realizar los lanzamientos y estimar 
crosses = 0
pi_estimates = []

for i in range(1, num_trials + 1):
    crosses += throw_needle(L, D)
    pi_estimate = (2 * L * i) / (D * crosses) if crosses > 0 else 0
    pi_estimates.append(pi_estimate)

# Graficar la convergencia de la estimacin de 
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, label='Estimacin de ')
plt.axhline(y=np.pi, color='r', linestyle='-', label=' verdadero')
plt.xlabel('Nmero de lanzamientos')
plt.ylabel('Estimacin de ')
plt.title('Convergencia de la estimacin de  usando el experimento de la aguja de Buffon')
plt.legend()
plt.grid(True)
plt.show()
```
