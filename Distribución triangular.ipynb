```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del triángulo
a = 2  # Vertice izquierdo
b = 9  # Altura
c = 6  # Vertice derecho

# Generar números aleatorios con distribución triangular
num_samples = 100
random_numbers = np.random.triangular(left=a, mode=c, right=b, size=num_samples)

# Graficar el histograma de los números generados
plt.hist(random_numbers, bins=30, density=True, alpha=0.6, color='g')
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(random_numbers)
```

```python
plt.step(X,Y)
```

```python
plt.plot(X,Y)
plt.show()
```
