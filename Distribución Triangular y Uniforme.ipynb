```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x)
def f(x):
    if 0 <= x < 1:
        return (5/4) - x
    elif 1 <= x <= 2:
        return 1/4

# Generar puntos siguiendo la distribución f(x)
num_points = 25
generated_points = []

for _ in range(num_points):
    u = np.random.uniform(0, 2)
    y = f(u)
    generated_points.append(u)

plt.plot(generated_points)
```

```python

plt.hist(generated_points)
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python

X,Y=cdf(generated_points)
```

```python
plt.step(X,Y)
```

```python

plt.plot(X,Y)
plt.show()
```
