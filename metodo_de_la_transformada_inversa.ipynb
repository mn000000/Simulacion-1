```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetro para la distribución exponencial (tasa lambda)
lambda_param = 1.0

# Función de densidad de probabilidad (fdp) de la distribución exponencial
def exponential_pdf(x, lambda_param):
    return lambda_param * np.exp(-lambda_param * x)

# Función de distribución acumulativa (FDA) de la distribución exponencial
def exponential_cdf(x, lambda_param):
    return 1 - np.exp(-lambda_param * x)

# Generación de la muestra usando el método de la transformada inversa
def inverse_transform_sampling(lambda_param, size=1000):
    u = np.random.uniform(0, 1, size)
    return -np.log(1 - u) / lambda_param

# Generar muestra
sample = inverse_transform_sampling(lambda_param)

```

```python
# Graficar fdp
x = np.linspace(0, 10, 1000)
plt.figure(figsize=(12, 6))
plt.plot(x, exponential_pdf(x, lambda_param), label='fdp Exponencial(lambda=1)')
plt.title('Función de Densidad de Probabilidad (fdp)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python

# Graficar FDA
plt.figure(figsize=(12, 6))
plt.plot(x, exponential_cdf(x, lambda_param), label='FDA Exponencial(lambda=1)', color='green')
plt.title('Función de Distribución Acumulativa (FDA)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()
```

```python

# Graficar histograma de la muestra
plt.figure(figsize=(12, 6))
plt.hist(sample, bins=30, density=True, alpha=0.6, color='g')
plt.plot(x, exponential_pdf(x, lambda_param), 'r-', lw=2, label='fdp Exponencial(lambda=1)')
plt.title('Histograma de la Muestra')
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid()
plt.show()
```

```python
# Documentación adicional con notas y aclaraciones
print("""
Notas:
1. El método de la transformada inversa es una técnica de generación de números aleatorios que se utiliza cuando se conoce la función de distribución acumulativa (FDA) de la distribución objetivo.
2. En este ejemplo, la distribución objetivo es la distribución exponencial con un parámetro de tasa lambda=1.
3. La muestra se genera transformando una muestra uniforme en [0,1] usando la inversa de la FDA de la distribución exponencial.
4. La muestra generada se ajusta a la distribución objetivo, como se puede observar en el histograma comparado con la fdp de la distribución exponencial.
5. La función de distribución acumulativa (FDA) proporciona una visión general de cómo se distribuyen las probabilidades acumuladas a lo largo del rango de valores posibles.
""")
```
