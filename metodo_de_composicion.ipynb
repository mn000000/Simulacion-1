```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetros de la distribución exponencial
lambda_param = 1.0  # Tasa (lambda) de la distribución exponencial

# Generar una muestra de la distribución exponencial usando el método de composición
def compositional_method(sample_size, lambda_param):
    U = np.random.uniform(0, 1, sample_size)
    X = -np.log(1 - U) / lambda_param
    return X

# Tamaño de la muestra
sample_size = 10000

# Generar la muestra
sample = compositional_method(sample_size, lambda_param)

# Graficar la función de densidad de probabilidad (fdp)
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/lambda_param)

plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='Función de Densidad de Probabilidad (fdp)', color='blue')
plt.title('Función de Densidad de Probabilidad de la Distribución Exponencial')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar la función de distribución acumulativa (cdf)
cdf = expon.cdf(x, scale=1/lambda_param)

plt.figure(figsize=(12, 6))
plt.plot(x, cdf, label='Función de Distribución Acumulativa (cdf)', color='green')
plt.title('Función de Distribución Acumulativa de la Distribución Exponencial')
plt.xlabel('x')
plt.ylabel('Distribución Acumulativa')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar el histograma de la muestra generada
plt.figure(figsize=(12, 6))
plt.hist(sample, bins=50, density=True, alpha=0.6, color='gray', edgecolor='black', label='Histograma de la Muestra')
plt.plot(x, pdf, label='Función de Densidad de Probabilidad (fdp)', color='blue')
plt.title('Histograma de la Muestra y la Función de Densidad de Probabilidad')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
