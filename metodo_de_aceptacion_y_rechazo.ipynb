```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parámetros para la distribución Beta
a, b = 2, 5

# Función de densidad de probabilidad (fdp) de la distribución Beta
def beta_pdf(x, a, b):
    return beta.pdf(x, a, b)

# Función acumulativa de la distribución Beta
def beta_cdf(x, a, b):
    return beta.cdf(x, a, b)

# Generación de la muestra usando el método de aceptación y rechazo
def acceptance_rejection_sampling(pdf, c, size=1000):
    samples = []
    while len(samples) < size:
        x = np.random.uniform(0, 1)
        y = np.random.uniform(0, c)
        if y <= pdf(x):
            samples.append(x)
    return np.array(samples)

# Cálculo de la constante c (el valor máximo de la fdp)
x = np.linspace(0, 1, 1000)
c = max(beta_pdf(x, a, b))

# Generar muestra
sample = acceptance_rejection_sampling(lambda x: beta_pdf(x, a, b), c)
```

```python

# Graficar fdp
plt.figure(figsize=(12, 6))
plt.plot(x, beta_pdf(x, a, b), label='fdp Beta(2,5)')
plt.title('Función de Densidad de Probabilidad (fdp)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python

# Graficar FDA
plt.figure(figsize=(12, 6))
plt.plot(x, beta_cdf(x, a, b), label='FDA Beta(2,5)', color='green')
plt.title('Función de Distribución Acumulativa (FDA)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid()
plt.show()
```

```python

# Graficar histograma de la muestra
plt.figure(figsize=(12, 6))
plt.hist(sample, bins=30, density=True, alpha=0.6, color='g')
plt.plot(x, beta_pdf(x, a, b), 'r-', lw=2, label='fdp Beta(2,5)')
plt.title('Histograma de la Muestra')
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid()
plt.show()
```

```python
# Documentación adicional con notas y aclaraciones
print("""
Notas:
1. El método de aceptación y rechazo es una técnica de generación de números aleatorios que se utiliza cuando es difícil generar directamente muestras de una distribución objetivo.
2. En este ejemplo, la distribución objetivo es la distribución Beta con parámetros a=2 y b=5.
3. La constante c se elige como el valor máximo de la función de densidad de probabilidad (fdp) de la distribución objetivo para asegurar que todas las muestras sean aceptadas con una probabilidad menor o igual a 1.
4. La muestra generada se ajusta a la distribución objetivo, como se puede observar en el histograma comparado con la fdp de la distribución Beta.
5. La función de distribución acumulativa (FDA) proporciona una visión general de cómo se distribuyen las probabilidades acumuladas a lo largo del rango de valores posibles.
""")
```
