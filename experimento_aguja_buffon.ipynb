```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del experimento
L = 1.0  # Longitud de la aguja
D = 1.0  # Distancia entre las líneas
num_trials = 10000  # Número de lanzamientos

def throw_needle(L, D):
    # Generar un ángulo aleatorio entre 0 y π/2
    theta = np.random.uniform(0, np.pi/2)
    # Generar una distancia aleatoria del centro de la aguja a la línea más cercana
    y = np.random.uniform(0, D/2)
    # Determinar si la aguja cruza una línea
    if y <= (L/2) * np.sin(theta):
        return 1
    else:
        return 0

# Realizar los lanzamientos y estimar π
crosses = 0
pi_estimates = []

for i in range(1, num_trials + 1):
    crosses += throw_needle(L, D)
    pi_estimate = (2 * L * i) / (D * crosses) if crosses > 0 else 0
    pi_estimates.append(pi_estimate)

# Graficar la convergencia de la estimación de π
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, label='Estimación de π')
plt.axhline(y=np.pi, color='r', linestyle='-', label='π verdadero')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Estimación de π')
plt.title('Convergencia de la estimación de π usando el experimento de la aguja de Buffon')
plt.legend()
plt.grid(True)
plt.show()
```
